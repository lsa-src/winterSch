# 조건문
 - 프로그램의 실행 흐름을 바꾸고자 할 때 사용하는 제어문
 - 사용자가 지정한 조건에 따라 실행할 문장을 결정함

# if ... : elif ... :
- if 조건문 뒤에는 반드시 콜론(:)
- 바로 아래 문장부터 if문에 속하는 모든 문장에 들여쓰기(indentation)를 해야 함
- 파이썬에는 다른 언어에 있는 switch 문이 존재하지 않음 : if...elif...elif... 문으로 수행
- pass  : def 문이나 if 문처럼 코드 블록을 본문으로 갖는 표현에서 본문을 비워 둘 때 사용


```
# 조건문
# 짝수 홀수 구분 
x = int(input('값입력 : '))

if x % 2 == 0 :
    print(f'{x}는 짝수')

if x % 2 == 0 :
    print(f'{x}는 짝수')
else :
    print(f'{x}는 홀수')
```

```
# 성적처리 
x = float(input('성적입력 : '))

if x >= 70 :
    print(f'{x}는 Pass')
elif x >= 60 :
    print(f'{x}는 재시험')
else :
    print(f'{x}는 fail')
```


### 비교연산자 
+ < ,  >,  >=, <=, == , !=

### 논리연산자
+ and, or, not


```
x = 10
y = 20
```

```
#비교연산자 
print(f'{x} > {y} = { x > y }')
print(f'{x} < {y} = { x < y }')
print(f'{x} == {y} = { x== y }')
print(f'{x} != {y} = { x!= y }')
```

```
#논리연산자
print(f'{x > 10 and y > 10}')
print(f'{x > 10 or y > 10}')
print(f'{not x > 10 }')
```

# 멤버십 연산자 : in
+ 문자열이나 리스트나 튜플과 같이 연속적인 자료구조에 속한 멤버를 확인하기 위한 연산자

```
str = '안녕하세요.'
'안' in str
```

```
lst = list(range(5))
5 in lst
```

```
d = {'a' : 1, 'b' : 2}
'a' in d
```

```
2 in d.values()
```

# 반복문 - for
+ 컬렉션으로부터 하나씩 요소(element)를 가져와, 루프 내의 문장들을 실행

```
#문자열 
for item in str : print(item)
```

```
#리스트
for item in lst : print(item)
```

```
#딕션너리
for item in d : print(item)
```

```
for item in d.values() : print(item)
```

```
for key, value in d.items() : print(key, value)
```

# range() 함수
### **range(시작숫자, 종료숫자, step)**
+ 결과는 시작숫자부터 종료숫자 바로 앞 숫자까지 컬렉션 생성
+ 값을 확인하기 위해서는 순서가 있는 리스트나 튜플 컬렉션으로 변환해야 함
+ 시작숫자를 생략하면 0부터 생성
+ step을 생략하면 1씩 증가

```
xlist = range(10)
print(type(xlist))
print(xlist)

for item in xlist:
    print(item, end = ' ')
```

```
xlist = list(range(10))
print(type(xlist))
print(xlist)

for item in xlist:
    print(item, end = ' ')
```

## **For문을 이용한 리스트 변형**

```
xlist = ['1','2']
xlist = [int(item) for item in xlist]
print(xlist)
```

## **반복문 while**
+ while 키워드 다음의 조건식이 참일 경우 계속 while 안의 블럭을 실행

```
i = 1
while i <= 5 :
    print(i)
    i = i + 1
```

### **break 문**
+ 반복문 안에서 루프를 빠져나오기 위해 을 사용

### **continue문**
+ 루프 블럭의 나머지 문장들을 실행하지 않고 다음 루프로 직접 돌아가게 함 

```
i = 0
while i <= 5 :
    i = i + 1
    if i == 2 : continue
    
    print(i)
    
    if i == 4 : break
```
